# Tarea 2 de Optimización
Utilizando gradiente descendente, newton y gradiente descendente encontrar los mínimos de diferentes funciones. 

```python
import numpy as np

class Solvers():
    def __init__(self, gradient, matriz_hessiana=np.nan, dif_finite=np.nan, solver="gd", 
                 alpha=0.5, epsilon=np.nan, tol=1e-3, max_iter=1000):
        self.gradient = gradient
        self.matriz_hessiana = matriz_hessiana
        self.alpha = alpha
        self.tol = tol
        self.max_iter = max_iter
        self.solver = solver
        self.epsilon = epsilon
        self.dif_finite = dif_finite
        
    def __gd_solver(self, start_point):
        grad_ = self.gradient(start_point)
        print(grad_)
        convergence = np.dot(grad_, grad_)
        stop = (convergence < self.tol)
        iter_ = 0
        point = start_point
        while not stop:
            iter_ += 1
            point = point - self.alpha*grad_
            print("iteration", str(iter_), ", update point ", 
                  str(point[0]), str(point[1]))
            grad_ = self.gradient(point)
            print(grad_)
            stop = ((np.dot(grad_, grad_)) < self.tol) or \
                (iter_>self.max_iter)
        return point 
    
    def __newton(self, start_point):
        grad_ = self.gradient(start_point)
        print(grad_)
        hessiana_ = self.matriz_hessiana(start_point)
        print(hessiana_)
        convergence = np.dot(grad_, grad_)
        stop = (convergence < self.tol)
        iter_ = 0
        point = start_point
        while not stop:
            iter_ += 1
            print(grad_@np.linalg.inv(hessiana_))
            point = point-grad_@np.linalg.inv(hessiana_)
            print("iteration", str(iter_), ", update point ", 
                  str(point[0]), str(point[1]))
            grad_ = self.gradient(point)
            print(grad_)
            hessiana_ = self.matriz_hessiana(point)
            print(hessiana_)
            stop = ((np.dot(grad_, grad_)) < self.tol) or \
                (iter_>self.max_iter)
        return point 
    
    def __gd_finite_solver(self, start_point):
        grad_ = self.dif_finite(start_point, self.epsilon)
        print(grad_)
        convergence = np.dot(grad_, grad_)
        stop = (convergence < self.tol)
        iter_ = 0
        point = start_point
        while not stop:
            iter_ += 1
            point = point - self.alpha*grad_
            print("iteration", str(iter_), ", update point ", 
                str(point[0]), str(point[1]))
            grad_ = self.dif_finite(point, self.epsilon)
            print(grad_)
            stop = ((np.dot(grad_, grad_)) < self.tol) or \
                    (iter_>self.max_iter)
        return point 
    
    def solve(self, start_point):
        if self.solver=="gd":
            return self.__gd_solver(start_point)
        if self.solver=="nt":
            return self.__newton(start_point)
        if self.solver=="gd_finite":
            return self.__gd_finite_solver(start_point)
        else: 
            raise ValueError('Undefined solver')
```

### function 1:
$$f(\overrightarrow{x})=10(x_2-x_1^2)+(1-x_1)^2$$

#### a) Gradiente descendente

```python
def function_one(x):
    return np.array([-40*x[0]*(x[1]-x[0]**2)-2*(1-x[0]),
                    20*(x[1]-x[0]**2)])

initial = np.array([0,0])

sol_function_one = Solvers(gradient=function_one, alpha=0.01, max_iter=1000)
print(sol_function_one.solve(initial))
```

#### b) Netwon's Method

```python
def hessiana_one(x):
    return np.array([[-40*x[1]+120*(x[0]**2)+2, -40*x[0]],
                    [-40*x[0], 20]])

sol_function_one = Solvers(gradient=function_one, max_iter=5,
                          matriz_hessiana=hessiana_one, solver="nt")
print(sol_function_one.solve(initial))
```

#### c) Gradient descent finite difference

```python
def diference_finite_one(x, epsilon):
    return np.array([(10*(x[1]-(x[0]+epsilon)**2)**2+(1-x[0]-epsilon)**2-(10*(x[1]-(x[0])**2)**2+(1-x[0])**2))/epsilon,
                    (10*(x[1]+epsilon-(x[0])**2)**2+(1-x[0])**2-(10*(x[1]-(x[0])**2)**2+(1-x[0])**2))/epsilon])

sol_function_one = Solvers(gradient=np.nan, alpha=0.01, max_iter=500, dif_finite=diference_finite_one, 
                           epsilon=.05, solver="gd_finite")
print(sol_function_one.solve(initial))
```

## Regression Lasso

```python
import itertools 
import pandas as pd
from cvxopt import matrix, solvers

def expand_grid(data_dict):
    """Crear la matriz de las combinaciones de A.
    
    Args:
        data_dict: (p, 2) diccionario con los vectores (-1,1)
    
    return:
        pd.Datafram: (p, 2) matriz A.
    """
    rows = itertools.product(*data_dict.values())
    return pd.DataFrame.from_records(rows, columns=data_dict.keys())
```

```python
def regression_lasso_qp(X, y, t):
    """
    Estimacion de beta para una regression con regularizacion l1.
    
    Args:
        X: (n,p) np.array, los datos del problema
        y: (n) np.array, variable respuesta
        t: parametro t de regularizacion
        
    return:
        sol["x"]: solución optima encontrada
    """
    n, p = X.shape # tamano de los datos
    
    D = matrix(X.T@X, tc="d") # calculamos la matriz D
    q = matrix(2*y.T@X, tc="d") # calculamos el vector q
    
    a = np.array([-1,1]) # calculamos la matriz A
    aux = {}
    for i in range(p):
        aux["Var{}".format(i)] = a
    A = matrix(np.array(expand_grid(aux)), tc="d")
    
    b0 = matrix(np.repeat(t, 2**p).reshape((2**p)), tc="d") # vector b
    
    sol = solvers.qp(D, q, A, b0) # resolvemos el PQ.
    return print(sol["x"])
```

```python
from sklearn.datasets import load_boston # cargamos los datos
X, y = load_boston(return_X_y=True)
```

```python
regression_lasso_qp(X[:,0:2], y, 10)
```
