```python
import pandas as pd
import numpy as np
```

```python
def comprimir_por_categoria (centro, df):
    aux_df = df[df["centro_turístico"] == centro].copy()
    col_names = aux_df.columns
    col_names_por_estrella = []
    for col_name in col_names[2:]:
        for i in range(1,6,1):
            sufijo = "_E"+str(i)
            col_names_por_estrella.append(col_name+sufijo)

    new_df = pd.DataFrame(columns=col_names_por_estrella)


    #añadimos datos para 1 estrella:
    
    try:
        aux_df_E1 = aux_df[aux_df["categoria"] == "estrella_1"].copy().reset_index()

        new_names_E1 = [name + "_E1" for name in col_names[2:]]

        valores_E1 = aux_df_E1.loc[0,col_names[2:]]

        new_df.loc[0,new_names_E1] = list(valores_E1)
    except:
        print("E1_error, ", centro)


    #añadimos datos para 2 estrella:
    try:
        
        aux_df_E2 = aux_df[aux_df["categoria"] == "estrellas_2"].copy().reset_index()

        new_names_E2 = [name + "_E2" for name in col_names[2:]]

        valores_E2 = aux_df_E2.loc[0,col_names[2:]]

        new_df.loc[0,new_names_E2] = list(valores_E2)
    except:
        print("E2_error, ", centro)


    #añadimos datos para 3 estrellas:
    try:
        aux_df_E3 = aux_df[aux_df["categoria"] == "estrellas_3"].copy().reset_index()

        new_names_E3 = [name + "_E3" for name in col_names[2:]]

        valores_E3 = aux_df_E3.loc[0,col_names[2:]]

        new_df.loc[0,new_names_E3] = list(valores_E3)
    except:
        print("E3_error, ", centro)

    #añadimos datos para 4 estrellas:
    try:
        aux_df_E4 = aux_df[aux_df["categoria"] == "estrellas_4"].copy().reset_index()

        new_names_E4 = [name + "_E4" for name in col_names[2:]]

        valores_E4 = aux_df_E4.loc[0,col_names[2:]]

        new_df.loc[0,new_names_E4] = list(valores_E4)
    except:
        print("E4_error, ", centro)

    #añadimos datos para 5 estrella:
    try:
        aux_df_E5 = aux_df[aux_df["categoria"] == "estrellas_5"].copy().reset_index()

        new_names_E5 = [name + "_E5" for name in col_names[2:]]

        valores_E5 = aux_df_E5.loc[0,col_names[2:]]

        new_df.loc[0,new_names_E5] = list(valores_E5)
    except:
        print("E5_error, ", centro)

    new_df["centro_turistico"] = centro

    return new_df.set_index("centro_turistico")
```

```python
centros_turisticos = pd.read_csv("hotel_categoria_cuartos_turistas.csv")
centros_turisticos = centros_turisticos.drop("corredor_turístico", axis = 1)



centros_nombres = list(set(centros_turisticos["centro_turístico"]))

centros_df_lista = []

for centro in centros_nombres:
    pass
    try:
        centros_df_lista.append(comprimir_por_categoria(centro, centros_turisticos))
    except:
        print("error, iteración " + centro)
    
    


df = centros_turisticos
centro = "Salamanca"
comprimir_por_categoria(centro, df)
centros_limpios = pd.concat(centros_df_lista)
```

```python

# agregamos 0's a NaN's
centros_limpios = centros_limpios.fillna(0)

centros_limpios
```

```python
centros_limpios.to_csv("centros_turisticos_limpios.csv")
```
