```python
import seaborn as sns

iris = sns.load_dataset("iris")
iris.head(5)
```

```python
sns.set()
sns.pairplot(iris, hue="species", height=2)
```

```python
sns.relplot(x="sepal_width", y="petal_width", hue="species", 
            data=iris, height=10)
```

# Silplifiquemos el problema.

Unimos las categorias 'versicolor' y 'virginica' en la clase 'vers_color'.

```python
iris2 = iris
spec2 = iris2['species'].copy()
spec2[spec2.eq('versicolor') | spec2.eq('virginica')] = 'vers_color'

```
