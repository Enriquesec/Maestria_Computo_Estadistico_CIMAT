Analisis exploratorio de datos pareados.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
datos_pareados = pd.read_csv("../Datos/ConjuntoParesDatos.csv")
```

```python
datos_pareados.head(2)
```

```python
datos_pareados.shape
```

```python
datos_pareados.info()
```

```python
for i in range(1,14):
    print("La correlación de ",i,np.corrcoef(datos_pareados["x_"+str(i)], datos_pareados["y_"+str(i)])[1,0])
    print("La media de x_",i,np.mean(datos_pareados["x_"+str(i)]))
    print("La media de y_",i,np.mean(datos_pareados["y_"+str(i)]))
    sns.scatterplot(data=datos_pareados, x="x_"+str(i), y="y_"+str(i))
    plt.title("Conjunto "+str(i))
    plt.show()
```

El objetivo de esta practica es analizar un conjunto de datos para observar que algunas veces (o muchas) no existe una relación lineal. Y los estadísticos clásicos como media, varianza y correlación son muy parecidas pero al graficarlas observamos una clara dependencia.
